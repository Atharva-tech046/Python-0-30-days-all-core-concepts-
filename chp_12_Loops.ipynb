```python
# BREAK AND CONTINUE 

# BREAK - It is used to terminate a loop when encountered

#CONTINUE - Terminates execution in current iteration and continues execution of the loop 
#with the next iteration .



# BREAK 
nums = (1,4,9,16,25,36,5,64,81,100,36)
i = 0
while i<=5:
    print(i)
    if(nums[i]==3):
        break
    else:
        print("Finding....")
    i+=1
print("END OF THE LOOP ")

```

```python
# BREAK AND CONTINUE 

#BREAK = used to terminate a loop when encountered 
#CONTINUE = terminates execution in current iteration and continues execution of loop in other iteration 
#continue is basically a skip statement used to skip the loop or a specific iteration 
```

```python
# eg break 

i = 0 

while i<=5:
    print(i)
    if i==3:
        break
    else:
        print("Finding ........")
    i+=1

print("Loop ended!")
```

```python
# eg continue 

i = 0 

while i<=5:
    if i==3:
        i+=1
        continue
    print(i)
    i+=1
```

```python
# Range function =
# Range function returns a sequence of numbers , starting form 0 by default and increments,
# by 1 (by default) and stops before a specified number 
# range() - inside the bracket you write where u want the range to end 
```

```python
#eg code 

print(range(5))

# eg code 2
nums = (range(5))

for i in nums:
    print(i) # this prints all the elements in range 5
    
```

```python
# start,stop and step size condition 
# range(start,stop,step)
#step size = the incremental value (i.e if step = 2 then values will increase like 2,4,6....)
#eg code  (start,stop)


for i in range(5,10,2):
    print(i)
```
