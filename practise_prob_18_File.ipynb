```python
# Q2) WAP replace all the occurances of 'java' with 'python' in the above file 'practise.txt'

with open("practise.txt","r")as f:
    data = f.read()
    print(data)

new_data = data.replace("Java","Python")
print(new_data)

with open("practise.txt","w")as f:
    f.write(new_data)
    

```

```python
# Q3)WAP to find the occurence of the word 'learning' in in the file practise.txt
word = 'learning'
with open("practise.txt","r")as f:
    data = f.read()
    if (data.find(word) != -1):
        print("Found")
    else:
        print("Not found!")
```

```python
# Q4)WAF to find which line of the file does the word 'learning' if not found print -1 

def line_check():
    word = 'learning'
    data = True
    line_no = 1
    with open("practise.txt","r")as f:
        while data:
            data = f.readline()
            if(word in data):
                print(line_no)
            line_no += 1
line_check()
```

```python
# Q5) WAP for from a file containing numbers seperated by comma , printthe count of even numbers 
# makea file and pur numbers in it and perform operations on that file

with open("numbers.txt","r") as f:
    data = f.read()
    print(data)

num = ""
for i in range(len(data)):
    if(data[i]==","):
        print(num)
        num = "" 
    else:
        num += data[i]
        
```

```python
# NOW ANOTHER METHOD TO DO THIS PROBLEM IS , USING 'SPLIT' FUNCTION

count = 0 
with open("numbers.txt","r")as f:
    data = f.read()
    
    nums = data.split (",")
    for val in nums :
        if(int(val) % 2 == 0 ):
            count+=1

print(count)
```
