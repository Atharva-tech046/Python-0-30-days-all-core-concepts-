```python
# PRACTISE PROBLEM 

#Q1) Print the following elements using a for loop 
#[1,4,9,16,25,36,49,64,81,100]

nums = [1,4,9,16,25,36,49,64,81,100]

for val in nums:
    print(val)
```

```python
#Q2) Search for number x in the tuple using the loop :
#(1,4,9,16,25,36,49,64,81,100)

tup=(1,4,9,16,25,36,49,64,81,100)

x=36

for val in tup:
    if(val==x):
        print("Found!... Value of x =",val)
    else:
        print("Finding")


# this process is known as linear search 
```
