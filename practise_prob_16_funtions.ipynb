```python
#Q1 WAP to print length of the list (using fucntion and define paramenter)

cities = ["Mumbai","Pune","Los Angles","Dallas","Banglore"]

def len_list(list):
    print(len(list))
    
len_list(cities)
```

```python
#Q2 WAP to write the elements of a list in a single line (list parameters)

cars = ["Mercedes","Ferrari","Pagani","Bentley"]

def print_list(list):
    print(list)

print_list(cars)
```

```python
#Q2 can also be written as 

cars = ["Mercedes","Ferrari","Pagani","Bentley"]

def print_list(list):
    for item in list:
        print(item,end=" ")

print_list(cars)
```

```python
#Q3 WAP to print the factorial of n (n is the parameter)

def cal_fac(n):
    fact=1
    for i in range (1,n+1):
        fact *= i 
    print(fact)

cal_fac(6)

       
```

```python
#WAP to convert USD to INR

def conv_app(a):
    usd = 94
    conv = a*usd
    print("Entered amount in USD: $",a)
    print("Final amount in INR: ₹",conv)

conv_app(40)
```
