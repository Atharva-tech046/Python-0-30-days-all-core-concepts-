```python
#Q1) WAP to find sum of first n numbers (using while loop)

n = 10 
sum=0
i=1

while i<=n:
    sum += i
    i+=1
print("Total sum:  ",sum)
```

```python
#Q2) WAP to find factorial for first n natural numbers 


n = 10 
prod = 1
i=1
for i in range(i,n):
    prod *= i
    i+= 1

print("Factorial of first n numbers :   ",prod)

```

```python
# using for loop 

n = 10 

fact =1
i=1

for i in range (i,n+1):
    fact*=i
print(fact)
```
