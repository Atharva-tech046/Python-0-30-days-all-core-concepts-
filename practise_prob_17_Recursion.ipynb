```python
#Q1 Write a recursive function to calculate the sum of first n natural numbers . 

def cal_sum(n):
    if (n==0):     #-----> Base case 
        return 0 
    return cal_sum(n-1)+n  #---> Recursive function 

sum = cal_sum(5)   
print(sum)
```
