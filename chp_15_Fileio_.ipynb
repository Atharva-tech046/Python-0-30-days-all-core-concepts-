```python
# File I/O  = These are functions/ operations performed on a file using python
# These operations include read and write functions 
# Types of files we are dealing with here are: 
# Text files - .txt,.docx,.log....etc 
# Binary files - .mp4,jpeg,.....etc
```

```python
# Operations on a file :
# Open , read and close 

# syntax = {f= open("file name",(mode))}

# here mode refers to the operations to be performed 
#i.e read or write 

# eg syntax :
#data = f.read()
# f.close()
```

```python
# now accessing file using I/O fucntions (you can make any file of your choice , a .txt file)

f = open("demo.txt","r")
data = f.read()
print(data)
print(type(data))
f.close()
```

```python
# r = open for reading 
# w = open fro writing 
# x = creating a new file and opening it for writing 
# a = opening for writing and appending data or whatever info at the end 
# b = binary mode 
# t = text mode 
# + = open a disk file for updating (reading and writing)
```

```python
#READING A FILE 

#syntax info - 

# data = f.read() -------> reads entire info / string 
# data = f.readline() -------> reads one line at a time 

# in () data we can specify how many charac. do we want it to read 

#eg code

f = open("demo.txt","r")

data = f.read(5)
print(data)
print(type(data))
f.close()
```

```python
# now printing one line at a time 

f = open("demo.txt","r")

line1 = f.readline()
print(line1)

line2 = f.readline()
print(line2)

f.close()
```

```python
# WRITE MODE (w)

#eg code 

f = open("demo.txt","w")

f.write("Hey there! I am Atharva.") # ----> so basically write function overwrites the whole file 

f.close()
```

```python
# APPEND MODE (a)

#eg code 

f = open("demo.txt","a")

f.write("I love Python.") #------> Appending the string in demo.txt

f.close()
```

```python
# MODE FOR READING AND WRITING BOTH (r+)
# This mode will enable the user to read as well as write in the file without truncating
# truncate here means without deleting thr previous data / info 
  
f = open("demo.txt","r+")

f.write("This is due to 'r+' fucntion") #--------> This fucntion will overwrite from the starting 

f.close()
```

```python
# POINTS TO REMEMBER 

# r+ --> used for read and overwrite (pointer at the start) ..... no truncate 

# w+ --> used for write and overwriting ..... truncates (deletes all the previous info and overwrites)

# a+ --> used for read and append (pointer at the last)....... not truncate 
```

```python
# 'WITH' SYNTAX

# eg code 

with open("demo.txt","r") as f:
    data = f.read()
    print(data)
    print(type(data))
  
```
