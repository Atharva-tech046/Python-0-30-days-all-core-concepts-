```python
#RECURSION = When a function calls itself repeatedly 
# Recursion is similar to loops but used in specific cases only
```

```python
#eg code for recursive function 

def show(n):
    if(n==0):   #-----> Base case 
        return
    print(n)
    print(n-1)

show(5)
```

```python
#eg code 2 
#understanding recursion with the eg of factorial code 

def fact(n):
    if (n==0 or n==1):
        return 1
    else:
        return n*fact(n-1) #-----> Reccurance relation 
    
print(fact(4))
```
