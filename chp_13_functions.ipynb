```python
# FUNCTION - IT IS A BLOCK OF STATEMENT THAT PERFORMS A SPECIFIC TASK 
# How do we write a function ? ----- > we define it using 'def func_():'
#And while calling it we write it as func(param1,param2) [refers to paramaeter 1 and 2]
```

```python
#eg code 

def cal_sum(a,b):    #--------> fucntion , (a,b) are the parameters
    sum = a+b
    return(sum)
 
cal_sum(10,40)#-----------> caling the fucntion 
```

```python
#eg code 2 
#WAP to print avg of 3 numbers 

def cal_avg(a,b,c):
    sum = a+b+c
    avg = sum/3
    return(avg)

cal_avg(10,30,70)
```
